```python
import matplotlib.pyplot as plt
# Parse dates and create a datetime column
data['Date'] = pd.to_datetime(data['Rok'].astype(str) + '-' + data['Miesiac'].astype(str) + '-' + data['Dzien'].astype(str), format='%Y-%m-%d')

# Group by 'Nazwa Stacji' and find min/max dates for each station
station_date_range = data.groupby('Nazwa Stacji')['Date'].agg(['min', 'max'])

# Calculate the total possible days (from min to max date in the dataset)
total_days = (data['Date'].max() - data['Date'].min()).days + 1

# Calculate days covered for each station
station_date_range['Days Covered'] = (station_date_range['max'] - station_date_range['min']).dt.days + 1

# Calculate percentage of coverage
station_date_range['Coverage %'] = (station_date_range['Days Covered'] / total_days) * 100

# Plotting time range coverage for each station
fig, ax = plt.subplots(figsize=(10, 20))  # Increasing the vertical size of the plot
for index, row in station_date_range.iterrows():
    ax.plot([row['min'], row['max']], [index, index], marker='o', linestyle='-', label=index)
ax.set_yticks(range(len(station_date_range)))
ax.set_yticklabels(station_date_range.index)
ax.set_title('Time Range Coverage for Each Station')
ax.set_xlabel('Date')
ax.set_ylabel('Station')
plt.xticks(rotation=45)  # Rotating x-axis labels for better readability
plt.tight_layout()  # Adjust layout to fit everything nicely
plt.show()
```

```python
import pandas as pd

# Assuming df is already loaded and includes the 'Date' column as datetime
df['Date'] = pd.to_datetime(df['Date'])  # Ensure 'Date' is in datetime format

# Function to apply to each group
def fill_rainfall(group):
    # Calculate the rolling median within each station group for 'Suma Opadow [mm]'
    group['Rolling Median Rain'] = group['Suma Opadow [mm]'].rolling(window=14, center=True).mean()
    
    # Mask for where there's a missing measurement
    mask = (group['Status pomiaru SMDB'] == 8.0) & (group['Suma Opadow [mm]'] == 0.0)
    group.loc[mask, 'Suma Opadow [mm]'] = group.loc[mask, 'Rolling Median Rain']
    
    return group.drop(columns=['Rolling Median Rain'])  # Return the group without the extra column

# Apply the function to each station group
df = df.groupby('Nazwa Stacji').apply(fill_rainfall).reset_index(drop=True)

# Optionally, reset index if you need to use the DataFrame differently late
```

```python
# Convert the 'Date' column to datetime if it's not already
df['Date'] = pd.to_datetime(df['Date'])

# Filter the DataFrame to include only the rows with 'Date' from 1985 to 2015 inclusive
filtered_df = df[(df['Date'].dt.year >= 1985) & (df['Date'].dt.year <= 2015)]

# Now group by 'Nazwa Stacji' and filter out those stations that have data for each year in the range
stations_with_full_coverage = filtered_df.groupby('Nazwa Stacji').filter(lambda x: x['Date'].dt.year.nunique() == (2015-1985+1))

# Get the unique list of stations that have full coverage from 1985 to 2015
stations = stations_with_full_coverage['Nazwa Stacji'].unique()

# Print the list of stations
print(stations)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import gamma, norm

def calculate_spi_for_period(values):
    # Ensure that values are positive and not all zeros for fitting gamma distribution
    filtered_values = values[values > 0]
    
    if len(filtered_values) == 0:
        return pd.Series([np.nan] * len(values), index=values.index)
    
    # Fit the gamma distribution
    params = gamma.fit(filtered_values, floc=0)
    
    # Calculate the CDF from the gamma distribution
    cdf_values = gamma.cdf(values, *params)
    
    # Transform CDF to Z-scores (SPI)
    spi = norm.ppf(cdf_values)
    
    # Replace infinite values with NaNs to handle edge cases
    spi[np.isinf(spi)] = np.nan
    
    return pd.Series(spi, index=values.index)

def calculate_spi_adjusted_v4(data):
    if 'Date' not in data.columns or pd.to_datetime(data['Date'], errors='coerce').isna().any():
        data['Date'] = pd.to_datetime(data[['Rok', 'Miesiac', 'Dzien']])
    
    data.set_index('Date', inplace=True)
    grouped = data.groupby('Nazwa Stacji').resample('ME').sum()['Suma Opadow [mm]']
    
    # Creating a DataFrame to store the rolling sums
    spi_data = pd.DataFrame({
        'SPI-1': grouped.rolling(window=1, min_periods=1).sum(),
        'SPI-3': grouped.rolling(window=3, min_periods=3).sum(),
        'SPI-12': grouped.rolling(window=12, min_periods=12).sum()
    })

    # Applying the SPI calculation for each period
    for period in ['SPI-1', 'SPI-3', 'SPI-12']:
        spi_data[period] = calculate_spi_for_period(spi_data[period])
    
    return spi_data


spi_adjusted_data_v4 = calculate_spi_adjusted_v4(df)


```
