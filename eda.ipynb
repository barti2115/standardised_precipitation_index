```python
import matplotlib.pyplot as plt
# Parse dates and create a datetime column
data['Date'] = pd.to_datetime(data['Rok'].astype(str) + '-' + data['Miesiac'].astype(str) + '-' + data['Dzien'].astype(str), format='%Y-%m-%d')

# Group by 'Nazwa Stacji' and find min/max dates for each station
station_date_range = data.groupby('Nazwa Stacji')['Date'].agg(['min', 'max'])

# Calculate the total possible days (from min to max date in the dataset)
total_days = (data['Date'].max() - data['Date'].min()).days + 1

# Calculate days covered for each station
station_date_range['Days Covered'] = (station_date_range['max'] - station_date_range['min']).dt.days + 1

# Calculate percentage of coverage
station_date_range['Coverage %'] = (station_date_range['Days Covered'] / total_days) * 100

# Plotting time range coverage for each station
fig, ax = plt.subplots(figsize=(10, 20))  # Increasing the vertical size of the plot
for index, row in station_date_range.iterrows():
    ax.plot([row['min'], row['max']], [index, index], marker='o', linestyle='-', label=index)
ax.set_yticks(range(len(station_date_range)))
ax.set_yticklabels(station_date_range.index)
ax.set_title('Time Range Coverage for Each Station')
ax.set_xlabel('Date')
ax.set_ylabel('Station')
plt.xticks(rotation=45)  # Rotating x-axis labels for better readability
plt.tight_layout()  # Adjust layout to fit everything nicely
plt.show()
```

```python
import pandas as pd

# Assuming df is already loaded and includes the 'Date' column as datetime
df['Date'] = pd.to_datetime(df['Date'])  # Ensure 'Date' is in datetime format

# Function to apply to each group
def fill_rainfall(group):
    # Calculate the rolling median within each station group for 'Suma Opadow [mm]'
    group['Rolling Median Rain'] = group['Suma Opadow [mm]'].rolling(window=14, center=True).mean()
    
    # Mask for where there's a missing measurement
    mask = (group['Status pomiaru SMDB'] == 8.0) & (group['Suma Opadow [mm]'] == 0.0)
    group.loc[mask, 'Suma Opadow [mm]'] = group.loc[mask, 'Rolling Median Rain']
    
    return group.drop(columns=['Rolling Median Rain'])  # Return the group without the extra column

# Apply the function to each station group
df = df.groupby('Nazwa Stacji').apply(fill_rainfall).reset_index(drop=True)

# Optionally, reset index if you need to use the DataFrame differently late
```

```python
import pandas as pd
import numpy as np
from scipy.stats import gamma, norm

def calculate_spi_for_period(values):
    # Ensure that values are positive and not all zeros for fitting gamma distribution
    filtered_values = values[values > 0]
    
    if len(filtered_values) == 0:
        return pd.Series([np.nan] * len(values), index=values.index)
    
    # Fit the gamma distribution
    params = gamma.fit(filtered_values, floc=0)
    
    # Calculate the CDF from the gamma distribution
    cdf_values = gamma.cdf(values, *params)
    
    # Transform CDF to Z-scores (SPI)
    spi = norm.ppf(cdf_values)
    
    # Replace infinite values with NaNs to handle edge cases
    spi[np.isinf(spi)] = np.nan
    
    return pd.Series(spi, index=values.index)

def calculate_spi_adjusted_v6(data):
    # if 'Date' not in data.columns or pd.to_datetime(data['Date'], errors='coerce').isna().any():
    #     data['Date'] = pd.to_datetime(data[['Rok', 'Miesiac', 'Dzien']])
    data['Date'] = pd.to_datetime(data['Date'])
    data.set_index('Date', inplace=True)
    data_grouped = data.groupby('Nazwa Stacji')

    # Prepare DataFrame to collect results
    results = pd.DataFrame()

    for name, group in data_grouped:
        # Perform rolling sum within each group
        grouped = group.resample('ME').sum()['Suma Opadow [mm]']
        spi_data = pd.DataFrame({
            'SPI-1': grouped.rolling(window=1, min_periods=1).sum(),
            'SPI-3': grouped.rolling(window=3, min_periods=3).sum(),
            'SPI-12': grouped.rolling(window=12, min_periods=12).sum()
        })

        # Calculate SPI for each period
        for period in ['SPI-1', 'SPI-3', 'SPI-12']:
            spi_data[period] = calculate_spi_for_period(spi_data[period])

        spi_data['Nazwa Stacji'] = name
        results = pd.concat([results, spi_data])

    # Adjust index and columns for output
    results.reset_index(inplace=True)
    results.set_index(['Nazwa Stacji', 'Date'], inplace=True)
    
    # Formatting Date to Year-Month only
    results.index = results.index.set_levels([results.index.levels[0], results.index.levels[1].to_period('M')])
    
    return results

# Example usage of the function with sample data (uncomment and fill the 'df' with your data)

# print(spi_adjusted_data_v6)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assume 'spi_adjusted_data' is loaded and available

# Statistical summary for a specific station
def station_summary(station_name):
    data = spi_adjusted_data.loc[station_name]  # Ensure 'spi_adjusted_data' is defined and has the station as the index
    print(data.describe())
    
    spi_types = spi_adjusted_data.columns  # Update 'spi_adjusted_data_v6' to 'spi_adjusted_data' if that's your DataFrame name
    fig, axes = plt.subplots(nrows=len(spi_types), figsize=(12, 6 * len(spi_types)))
    
    for ax, spi_type in zip(axes, spi_types):
        # Using a simple line plot instead of errorbar
        ax.plot(data.index.to_timestamp(), data[spi_type], label=spi_type)
        ax.set_title(f'{spi_type} Time Series for {station_name}')
        ax.set_ylabel(f'{spi_type} Value')
        ax.set_xlabel('Date')
        ax.legend()

    # Ensure plt.suptitle() is called before plt.show() to avoid graphical issues
    plt.suptitle(f'Correlation Between SPI Variants for {station_name}')
    # Comparison of SPI variants using pairplot on cleaned data
    sns.pairplot(data.dropna())
    plt.show()

# Example call
station_summary('CISNA')
```

```python
# Statistical summary for all stations
def plot_all_stations(spi_data):
    grouped_data = spi_data.groupby(level=1)  # Group by Date
    mean_spi = grouped_data.mean()

    spi_types = mean_spi.columns
    fig, axes = plt.subplots(nrows=len(spi_types), figsize=(12, 6 * len(spi_types)))

    for ax, spi_type in zip(axes, spi_types):
        ax.plot(mean_spi.index.to_timestamp(), mean_spi[spi_type], label=spi_type)
        ax.set_title(f'Mean {spi_type} Line Plot for All Stations')
        ax.set_ylabel(f'Mean {spi_type} Value')
        ax.set_xlabel('Date')
        ax.legend()

    plt.tight_layout()
    plt.show()

```

```python
import folium
import pandas as pd

# Assuming 'data2' is your DataFrame with the 'Coordinates' column containing lists or strings of coordinates
# And 'spi_adjusted_data' is the DataFrame containing your latest SPI data with 'Nazwa Stacji' as one of the columns

# Parse the 'Coordinates' column
def parse_coordinates(coord):
    if isinstance(coord, str):
        # This is a string representation of a list, convert it to a list
        return [float(c) for c in coord.strip('[]').split(',')]
    elif isinstance(coord, list):
        # This is already a list, use it as is
        return coord
    else:
        # This is a numpy array or another type, convert to a list
        return coord.tolist()

# Apply the function to parse coordinates
data2[['longitude', 'latitude']] = data2['Coordinates'].apply(parse_coordinates).tolist()

# Now, let's merge the 'spi_adjusted_data' with 'data2'
latest_spi = spi_adjusted_data.groupby('Nazwa Stacji').last().reset_index()
merged_data = pd.merge(data2, latest_spi, on='Nazwa Stacji', how='right')

# Initialize a Folium map centered around the average coordinates
avg_lat = merged_data['latitude'].mean()
avg_lon = merged_data['longitude'].mean()
m = folium.Map(location=[avg_lat, avg_lon], zoom_start=6)

# Add circle markers to the map for each station with an SPI-1 value
for idx, row in merged_data.iterrows():
    folium.CircleMarker(
        location=(row['latitude'], row['longitude']),
        radius=5, # Radius of the circle marker
        popup=f"{row['Nazwa Stacji']}: SPI-1 {row['SPI-1']}", # Popup text
        color='blue', # Color of the marker
        fill=True,
        fill_color='blue' # Fill color of the marker
    ).add_to(m)

# Display the map
m

# Save to an HTML file if necessary
#m.save("/mnt/data/spi_map.html")

# Output the path to the saved HTML file so it can be downloaded
#"/mnt/data/spi_map.html"
```

```python
import folium
import pandas as pd
from ipywidgets import interact, Dropdown
from branca.colormap import LinearColormap
# Assuming 'spi_adjusted_data' and 'data2' are preloaded as described before
# spi_adjusted_data has 'SPI-1', 'SPI-3', 'SPI-12', 'Date', and 'Nazwa Stacji'
# data2 has 'Nazwa Stacji', 'Coordinates' (which we've already parsed into 'latitude' and 'longitude')

# Define a function to draw the map based on user selections
def draw_spi_map(year, month, spi_index):
    # Filter data by year and month
    filtered_data = spi_adjusted_data[
        (spi_adjusted_data.index.get_level_values('Date').year == year) &
        (spi_adjusted_data.index.get_level_values('Date').month == month)
    ]

    
    merged_data = pd.merge(data2, filtered_data[spi_index], on='Nazwa Stacji', how='right')

    
    # Initialize the Folium map centered around the mean location
    m = folium.Map(location=[merged_data['latitude'].mean(), merged_data['longitude'].mean()], zoom_start=6)
    
    spi_values = filtered_data[spi_index].dropna().values
    min_val, max_val = min(spi_values), max(spi_values)
    
    # Define a colormap
    clm = plt.get_cmap('plasma')
    colors = [clm(i) for i in range(clm.N)]
    folium_color_map = LinearColormap(colors, vmin=min_val, vmax=max_val).to_step(n=10)
    
    m.add_child(folium_color_map)
    
    # Add circle markers with popups

    for idx, row in merged_data.iterrows():
        value = row[spi_index]
        color=folium_color_map(value)
        folium.CircleMarker(
            location=(row['latitude'], row['longitude']),
            radius = 5,
            weight=1,
            color = color,# Marker size
            fill = True,
            fill_color = color,
            fill_opacity=1,
            popup=f"{row['Nazwa Stacji']}: {spi_index} {row[spi_index]}",  # Popup text
        ).add_to(m)
    
    # Display the map
    display(m)

# Dropdown widgets for user input
years = sorted(spi_adjusted_data.index.get_level_values('Date').year.unique())
months = sorted(spi_adjusted_data.index.get_level_values('Date').month.unique())
spi_indices = ['SPI-1', 'SPI-3', 'SPI-12']

year_dropdown = Dropdown(options=years, description='Year:')
month_dropdown = Dropdown(options=months, description='Month:')
spi_index_dropdown = Dropdown(options=spi_indices, description='SPI Index:')

# Interactive widget to draw the map based on selections
interact(draw_spi_map, year=year_dropdown, month=month_dropdown, spi_index=spi_index_dropdown)
```
